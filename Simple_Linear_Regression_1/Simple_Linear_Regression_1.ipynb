```python
import pandas as pd
import numpy as np
```

```python
teams=pd.read_csv("teams.csv")
```

```python
teams
```

```python
X=teams[["athletes","prev_medals"]]
y=teams[["medals"]]
```

```python
X
```

```python
y
```

```python
X["intercept"]=1
X=X[["intercept","athletes","prev_medals"]]
X
```

```python
X_T=X.T
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y
B
```

```python
B.index=X.columns
```

```python
predictions = X @ B
predictions
```

```python
SSR = ((y - predictions) ** 2).sum()
SSR
```

```python
TSS = ((y - y.mean()) ** 2).sum()
TSS
```

```python
R2 = 1 - (SSR / TSS)
```
